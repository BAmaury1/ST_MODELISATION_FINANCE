```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tau, n, n_rho = 100, 1000, 10
```

```python
sigma = 1.
```

```python
rho = np. linspace (0, 1., n_rho)
epsilon = np. random. normal (0, sigma, size= (tau,n,n_rho))
X= np. zeros((tau, n, n_rho))
```

```python
for t in range (1, tau) :
    X[t] = rho * X[t-1] + epsilon [t-1]
```

```python
# First realization of the 8th rho
plt.plot (X[:,0,7])
plt.xlabel('time' ,fontsize=16)
plt.ylabel ('$X t$')
plt.tick_params(labelsize=14)
```

```python
# Second realization of the Sth rho
plt.plot (X[:,1,7])
plt.xlabel('time' ,fontsize=16)
plt.ylabel('$X_t$' ,fontsize=16)
plt.tick_params (labelsize=14)
```

```python
# Second realization of the last rho
plt.plot (X[:,1,9])
plt.xlabel('time',fontsize=16)
plt.ylabel('$X_t$',fontsize=16)
plt.tick_params (labelsize=14)
```

```python
indices_time,indices_sim,indices_rho = np.where (np.sign (X[1]) != np. sign (X[1:]))
indices_time,indices_sim,indices_rho
```

```python
indices_pair,indices_first_occurrence = np.unique ([indices_sim, indices_rho],axis=1,return_index=True)
indices_pair,indices_first_occurrence
```

```python
stopping_time = np. full ( (n,n_rho) ,np.inf)
stopping_time [indices_pair [0],indices_pair[1]] = indices_time [indices_first_occurrence]
```

```python
plt.plot (rho, stopping_time.mean (axis=0),'-o') 
plt.xlabel(' ',fontsize=16)
plt.ylabel('stopping time' ,fontsize=16)
plt.tick_params(labelsize=14)
plt.grid()
plt.show ()
```
